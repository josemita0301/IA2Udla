## Introducción

Este proyecto tiene como objetivo analizar las palabras seleccionadas por los usuarios para describir la usabilidad de una aplicación móvil, con el fin de identificar temas y patrones comunes en estas descripciones. A través de la construcción de una red de documentos y la aplicación de algoritmos de detección de comunidades, buscamos agrupar palabras que los usuarios perciben de manera similar en diferentes aspectos de la usabilidad.

Se utilizaron varias técnicas, incluyendo la construcción de una matriz documento-término (DTM) y la medición de similitudes entre documentos utilizando métricas como la similitud de Jaccard y la distancia euclidiana. La representación gráfica de la red de palabras y la aplicación de algoritmos de detección de comunidades nos permitió obtener agrupaciones coherentes de términos, facilitando la interpretación de los resultados en el contexto de la usabilidad de la aplicación.

## Experimentación

### 1. Generación de la Matriz Documento-Término
En esta primera fase, el código procesa un conjunto de datos con palabras seleccionadas por usuarios para describir la usabilidad de una aplicación móvil. Usamos estas palabras para construir una matriz documento-término (DTM), donde cada fila representa un usuario (documento) y cada columna representa una palabra única. Esta matriz permite visualizar la presencia o ausencia de cada palabra en los documentos, generando una estructura que facilita análisis posteriores sobre la frecuencia de términos y patrones comunes entre los usuarios.



```python
# Import the libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_score
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import euclidean_distances
import networkx as nx
import community as community_louvain

# Import the CSV file into a pandas DataFrame
url = "https://raw.githubusercontent.com/marsgr6/estadistica-ux/main/data/words_ux.csv"
df = pd.read_csv(url)

# Combine words in each row into a single document
df['document'] = df['Words'].apply(lambda x: ' '.join(x.split(',')))

# Create a CountVectorizer object
vectorizer = CountVectorizer()

# Fit the vectorizer to the documents and transform them into a document-term matrix
dtm = vectorizer.fit_transform(df['document'])

# Convert the document-term matrix to a pandas DataFrame
dtm_df = pd.DataFrame(dtm.toarray(), columns=vectorizer.get_feature_names_out())

plt.figure(figsize=(10, 8))  # Adjust figure size if needed
sns.heatmap(dtm_df, cmap="gray", annot=False, cbar=False)  # Updated cmap and cbar
plt.title("Document-Term Matrix Heatmap")
plt.xlabel("Words")
plt.ylabel("Documents/Users")
plt.show()
```

<div style="border: 1px solid #007BFF; color: #000000; background-color: #E8F4FF; padding: 10px; border-radius: 5px; margin-top: 10px;">
  <strong style="color: #007BFF;">Información:</strong> Como resultado del código se muestra un mapa de calor de la matriz documento-término, donde cada cuadro blanco indica que un usuario ha seleccionado una palabra específica, y los cuadros negros muestran su ausencia. En cuanto a la dispersión y concentración de cuadros blancos entre las columnas, estas muestran cómo varía la selección de palabras entre los usuarios.
</div>

## 2.1 Similarity Measure: Jaccard Similarity
Ahora calculamos la matriz de similitud de Jaccard para evaluar la similitud entre los conjuntos de palabras seleccionadas por cada usuario. Para cada par de documentos, el código calcula el puntaje de similitud y construye una matriz simétrica de similitud, donde las filas y columnas representan usuarios.

```python
# Calculate Jaccard similarity matrix
num_docs = dtm_df.shape[0]
jaccard_matrix = np.zeros((num_docs, num_docs))

for i in range(num_docs):
    for j in range(i + 1, num_docs):
        jaccard_matrix[i, j] = jaccard_score(dtm_df.iloc[i], dtm_df.iloc[j], average='binary')  # average='binary' for presence/absence
        jaccard_matrix[j, i] = jaccard_matrix[i, j]  # Ensure symmetry

# Visualize similarity matrix as heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(jaccard_matrix, annot=False)  # Set annot=False
plt.title("Jaccard Similarity Matrix Heatmap")
plt.xlabel("Documents/Users")
plt.ylabel("Documents/Users")
plt.show()
```

<div style="border: 1px solid #007BFF; color: #000000; background-color: #E8F4FF; padding: 10px; border-radius: 5px; margin-top: 10px;">
  <strong style="color: #007BFF;">Información:</strong> La imagen generada es un mapa de calor de la matriz de similitud de Jaccard entre los documentos (usuarios). Los valores más claros representan mayor similitud entre documentos, mientras que los valores oscuros indican menor o ninguna similitud.
</div>


### 2.1.1 Gráfico de Red de Documentos: Jaccard Similarity
En etsa parte, construimos un gráfico de red donde cada nodo representa una palabra única seleccionada por los usuarios, y los enlaces entre nodos representan la similitud de Jaccard entre documentos que contienen estas palabras. PAquí se calcularon conexiones entre palabras comunes en documentos similares, utilizando la matriz de similitud de Jaccard como peso en los enlaces loq ue nos permite visualizar la relación entre las palabras en función de su co-ocurrencia en los documentos, formando una estructura de red que ilustra patrones en la percepción de la usabilidad.

```python
# Create a graph object
graph = nx.Graph()

# Add nodes (words) to the graph
words = vectorizer.get_feature_names_out()  # Get unique words from the vectorizer
graph.add_nodes_from(words)

# Add edges based on similarity matrix
num_docs = jaccard_matrix.shape[0]
for i in range(num_docs):
    for j in range(i + 1, num_docs):
        if jaccard_matrix[i, j] > 0:  # Connect if similarity is greater than 0
            # Get the words present in both documents
            common_words = [word for word in words if dtm_df.iloc[i][word] > 0 and dtm_df.iloc[j][word] > 0]
            
            # Add edges between common words with weight as Jaccard similarity
            for word1 in common_words:
                for word2 in common_words:
                    if word1 != word2:
                        if graph.has_edge(word1, word2):
                            # Increase weight if edge already exists
                            graph[word1][word2]['weight'] += jaccard_matrix[i, j]
                        else:
                            # Add edge with weight
                            graph.add_edge(word1, word2, weight=jaccard_matrix[i, j])

# Visualize the graph with improved layout
plt.figure(figsize=(12, 10))

# Choose a better layout algorithm
pos = nx.spring_layout(graph, k=0.3, iterations=50, seed=42)  # Adjust k, iterations, seed for better spacing

# Draw the graph
nx.draw(graph, pos, with_labels=True, node_color="lightblue", edge_color="gray", font_size=8, node_size=500)
plt.title("Document Network Graph (Nodes as Words - Improved Layout)")
plt.show()
```

<div style="border: 1px solid #007BFF; color: #000000; background-color: #E8F4FF; padding: 10px; border-radius: 5px; margin-top: 10px;">
  <strong style="color: #007BFF;">Información:</strong> La imagen muestra un gráfico de red donde los nodos representan palabras seleccionadas por los usuarios y las conexiones entre nodos indican similitud entre documentos. Los nodos agrupados y conectados más densamente indican palabras que tienden a aparecer juntas en la percepción de usabilidad. Los nodos más aislados representan términos menos comunes o con menor co-ocurrencia.
</div>


### 2.1.2 Red de Documentos Filtrada: Jaccard Similarity
Aquí, aplicamos un filtro a la red de documentos generada previamente, eliminando edges con un peso inferior a un umbral específico para simplificar la visualización. El umbral nos permite conservar únicamente las conexiones más fuertes, representando así relaciones significativas entre palabras. Además, eliminamos nodos sin conexión para mantener solo aquellos que forman parte de comunidades facilitando una interpretación más clara de las relaciones entre palabras.

```python
# Calculate edge weights and determine threshold
edge_weights = [data['weight'] for _, _, data in graph.edges(data=True)]
threshold = np.percentile(edge_weights, 0)  # Adjust percentile as needed

# Remove edges below the threshold
edges_to_remove = [(u, v) for u, v, data in graph.edges(data=True) if data['weight'] < threshold]
graph.remove_edges_from(edges_to_remove)
# Remove unconnected nodes
graph.remove_nodes_from(list(nx.isolates(graph)))

# Visualize the trimmed graph with improved layout and removed unconnected nodes
plt.figure(figsize=(12, 10))  # Adjust figure size if needed
pos = nx.spring_layout(graph, k=0.3, iterations=50, seed=42)  # Adjust layout parameters
nx.draw(graph, pos, with_labels=True, node_color="lightblue", edge_color="gray", font_size=8, node_size=500)
plt.title("Trimmed Document Network Graph (Nodes as Words - Improved Layout)")
plt.show()
```

<div style="border: 1px solid #007BFF; color: #000000; background-color: #E8F4FF; padding: 10px; border-radius: 5px; margin-top: 10px;">
  <strong style="color: #007BFF;">Información:</strong> La imagen muestra el gráfico de red filtrado. Al reducir la cantidad de enlaces, el gráfico resalta las palabras y conexiones más relevantes, agrupando términos en comunidades más claras.
</div>

### 2.1.3 Comunidades en al red de documentos
En esta parte, aplicamos el algoritmo de modularidad de Louvain para detectar comunidades dentro de la red de documentos. Este algoritmo agrupa nodos en comunidades basadas en la densidad de sus conexionespero tenemos que ajustar la resolución para controlar el tamaño de las comunidades y les damos colores diferentes para distinguirlas mejor.

```python
# Apply Louvain Modularity algorithm
resolution = 5.0  # Adjust this value to control community size
partition = community_louvain.best_partition(graph, resolution=resolution)

# Get community colors
community_colors = [partition[node] for node in graph.nodes()]

# Visualize the graph with communities using a better layout
plt.figure(figsize=(12, 10)) 
pos = nx.spring_layout(graph)

# Draw nodes and edges
nx.draw_networkx_nodes(graph, pos, node_color=community_colors, cmap="viridis", node_size=500)
nx.draw_networkx_edges(graph, pos, edge_color="gray", alpha=0.5)  # Adjust alpha for edge visibility
nx.draw_networkx_labels(graph, pos, font_size=10)

# Apply Louvain Modularity algorithm
partition = community_louvain.best_partition(graph)

# Print community structure in desired format
community_words = {}  # Dictionary to store words for each community
for node, community_id in partition.items():
    if community_id not in community_words:
        community_words[community_id] = []
    community_words[community_id].append(node)

for community_id, words in community_words.items():
    print(f"Community {community_id}: {', '.join(words)}")

plt.title("Document Network with Communities (Jaccard)")
plt.show()
```

## 2.2 Similarity Measure: Euclidean Distance
Primero calculamos la matriz de distancia euclidiana para medir la similitud entre documentos en función de las palabras seleccionadas. A diferencia de la similitud de Jaccard, esta considera la proximidad en un espacio multidimensional. Después pasamos a convertir esta distancia en una métrica de similitud, donde valores cercanos a 1 indican alta similitud y valores cercanos a 0 indican baja similitud.

```python
# Calculate Euclidean distance matrix
euclidean_matrix = euclidean_distances(dtm_df)

# Convert to similarity matrix
similarity_matrix = 1 / (1 + euclidean_matrix)

# Convert to DataFrame (optional)
similarity_df = pd.DataFrame(similarity_matrix, index=dtm_df.index, columns=dtm_df.index)

# Plot heatmap
plt.figure(figsize=(10, 8))  # Adjust figure size if needed
sns.heatmap(similarity_df, annot=False, cmap="viridis")  # You can change the colormap
plt.title("Euclidean Distance-Based Similarity Matrix Heatmap")
plt.xlabel("Documents/Users")
plt.ylabel("Documents/Users")
plt.show()
```

</div>

### 2.2.1 Gráfico de Red de Documentos: Euclidean Distance
Se connstruye un gráfico de red donde cada nodo representa una palabra seleccionada por los usuarios, y los edges se basan en la similitud derivada de la distancia euclidiana entre documentos. Esta similitud refleja la cercanía de los documentos en un espacio multidimensional, donde documentos más cercanos tienen palabras en común con mayor frecuencia. 
Para simplificar el gráfico, aplicamos un umbral de peso, eliminando conexiones débiles y manteniendo solo las relaciones más significativas. Esto nos permite visualizar agrupaciones clave de palabras en función de su similitud en los documentos.

```python
# Create a graph object
graph = nx.Graph()

# Add nodes (words) to the graph
words = vectorizer.get_feature_names_out()  # Get unique words from the vectorizer
graph.add_nodes_from(words)

# Add edges based on similarity matrix
num_docs = similarity_matrix.shape[0]  # Use similarity_matrix instead of jaccard_matrix
for i in range(num_docs):
    for j in range(i + 1, num_docs):
        if similarity_matrix[i, j] > 0:  # Connect if similarity is greater than 0
            # Get the words present in both documents
            common_words = [word for word in words if dtm_df.iloc[i][word] > 0 and dtm_df.iloc[j][word] > 0]
            
            # Add edges between common words with weight as Euclidean similarity
            for word1 in common_words:
                for word2 in common_words:
                    if word1 != word2:
                        if graph.has_edge(word1, word2):
                            # Increase weight if edge already exists
                            graph[word1][word2]['weight'] += similarity_matrix[i, j]
                        else:
                            # Add edge with weight
                            graph.add_edge(word1, word2, weight=similarity_matrix[i, j])

# Visualize the graph (you can adjust layout and visualization parameters as before)
edge_weights = [data['weight'] for _, _, data in graph.edges(data=True)]
threshold = np.percentile(edge_weights, 0)  # Adjust percentile as needed

# Remove edges below the threshold
edges_to_remove = [(u, v) for u, v, data in graph.edges(data=True) if data['weight'] < threshold]
graph.remove_edges_from(edges_to_remove)

# Visualize the trimmed graph (you can adjust layout and visualization parameters as before)
plt.figure(figsize=(12, 10))
nx.draw(graph, with_labels=True, node_color="lightblue", edge_color="gray", font_size=8, node_size=500)
plt.title("Trimmed Document Network Graph (Nodes as Words - Euclidean Distance)")
plt.show()
```

Como podemos ver, existen palabras agrupadas en el centro del gráfico que son los términos comunes que tienden a aparecer juntos en varios documentos, mientras que los nodos más aislados representan términos menos comunes.

## Evaluación

Para evaluar los temas obtenidos con cada medida de similitud (Jaccard y distancia Euclidiana), consideramos la **coherencia de los temas** y su **interpretabilidad** en función de las palabras agrupadas en cada comunidad. A continuación, se presenta un análisis comparativo de ambos métodos:

### Similitud de Jaccard

- **Coherencia**: Los temas identificados con la similitud de Jaccard presentan una alta coherencia, ya que agrupan palabras que suelen aparecer juntas en los documentos. Esto se debe a que Jaccard mide la similitud en función de la presencia o ausencia de palabras, lo que ayuda a agrupar términos que los usuarios perciben como relacionados en cuanto a características de usabilidad.
- **Interpretabilidad**: Las comunidades de Jaccard son interpretables y reflejan temas claros, como funcionalidad, complejidad, innovación y tiempo de consumo. Cada grupo de términos parece capturar aspectos bien definidos de la usabilidad, lo que facilita la comprensión de las percepciones de los usuarios.

### Distancia Euclidiana

- **Coherencia**: La distancia Euclidiana también presenta comunidades coherentes, aunque en menor grado que Jaccard. Dado que considera la proximidad en un espacio multidimensional, las comunidades incluyen palabras con una distribución similar de frecuencia y cantidad. Esto da como resultado agrupaciones que reflejan temas, pero algunas comunidades pueden ser menos compactas o contener términos más variados.
- **Interpretabilidad**: Las comunidades basadas en distancia Euclidiana son interpretables, pero en algunos casos menos definidas en comparación con Jaccard. Aunque encontramos temas similares (funcionalidad, complejidad, entretenimiento, etc.), la Euclidiana tiende a capturar una relación de proximidad más amplia, lo que puede incluir palabras que no necesariamente representan un mismo concepto de usabilidad.

## Decisión Final

**Similitud de Jaccard** se considera la mejor medida de similitud en este análisis, ya que produjo temas más coherentes y claramente interpretables. La agrupación de palabras es más compacta y directa, facilitando la identificación de temas específicos relacionados con la usabilidad. Este enfoque permite una interpretación más clara de las percepciones de los usuarios, lo cual es crucial para comprender y mejorar la experiencia del usuario en función de sus descripciones.

## Recomendaciones

1. **Explorar Otras Medidas de Similitud**:
   - Además de Jaccard y la distancia Euclidiana, sería útil experimentar con otras métricas como **coseno** o **Manhattan** para evaluar cómo afectan a la agrupación de palabras y a la coherencia de los temas. Cada medida de similitud puede capturar relaciones de diferentes maneras, y esto podría enriquecer el análisis.

2. **Ajuste de Parámetros en el Algoritmo de Detección de Comunidades**:
   - En el algoritmo de Louvain, ajustar la **resolución** puede ayudar a obtener comunidades de diferentes tamaños y granularidades. Esto permite capturar temas amplios o más específicos según el objetivo del análisis. Probar con distintas resoluciones puede revelar nuevas estructuras de temas y mejorar la interpretabilidad.

3. **Incluir Técnicas de Preprocesamiento de Texto Más Avanzadas**:
   - Considerar técnicas de preprocesamiento como **lematización** o **eliminación de stopwords específicas** relacionadas con la usabilidad puede mejorar la precisión de los grupos. Esto ayudaría a reducir ruido y a enfocar el análisis en palabras clave más relevantes.
   - También se podrían explorar técnicas como **reducción de dimensionalidad** (e.g., PCA o t-SNE) para simplificar el espacio de términos y facilitar la agrupación.

4. **Visualización de Redes con Filtrado de Peso Adaptativo**:
   - En la visualización de redes, ajustar el umbral de peso en función de la medida de similitud utilizada puede proporcionar gráficos más representativos. Adaptar el filtro para capturar solo las conexiones más significativas (según la métrica) puede hacer que los temas sean más evidentes y ayudar a interpretar los resultados.
